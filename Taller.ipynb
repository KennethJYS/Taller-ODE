```python
def f(t, y):
    return y-t**2+1
a=0.01
b=2
y_t0=0.5
N=10
ys, ts, h=ODE_euler(a=a,b=b,f=f,y_t0=y_t0,N=N)
```

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt

y_real=lambda t: t/(1+log(t))
tts= np.linspace(a,b,100)
y_valores_reales=[y_real(t) for t in tts]

plt.plot(ts,ys,marker="o",linestyle=":",label="Método de Euler")
plt.plot(tts,y_valores_reales,marker=".",linestyle="--",label="Solución Real")
plt.xlabel("Tiempo")
plt.ylabel("Y")
plt.title("Comparación de la solución real y la aproximación")
plt.legend()
plt.show()
```
